The regulations are as follow
* male: height > 175cm
* male: weight > 60kg && < 80kg
* blood type needs to be match: A to A B to B, C to C

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.externals.six import StringIO   
from sklearn.tree import export_graphviz
import pydotplus
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(np.random.randn(50,4), columns=['Height', 'Weight', 'bloodtype', 'Constellation'])
# print(np.random.randint(0,9))
```

```python
height=[]
weight=[]
female_blood=[]
female_constell=[]
male_blood=[]
male_constell=[]
like = []


def CONS(fe_co, ma_co):
    if fe_co==1:
        if ma_co==9 or ma_co==11 or ma_co==7 or ma_co==10:
            return(1)
        else:return(0)
    elif fe_co==2: 
        if ma_co==7 or ma_co==8 or ma_co==5 or ma_co==10:
            return(1)
        else:return(0)
    elif fe_co==3: 
        if ma_co==11 or ma_co==9 or ma_co==6 or ma_co==10:
            return(1)
        else:return(0)
    elif fe_co==4: 
        if ma_co==8 or ma_co==12 or ma_co==5 or ma_co==10:
            return(1)
        else:return(0)
    elif fe_co==5: 
        if ma_co==7 or ma_co==8 or ma_co==4 or ma_co==10:
            return(1)
        else:return(0)
    elif fe_co==6: 
        if ma_co==12 or ma_co==10 or ma_co==4 or ma_co==10:
            return(1)
        else:return(0)
    elif fe_co==7: 
        if ma_co==2 or ma_co==8 or ma_co==9 or ma_co==10:
            return(1)
        else:return(0)
    elif fe_co==8: 
        if ma_co==9 or ma_co==2 or ma_co==4 or ma_co==10:
            return(1)
        else:return(0)
    elif fe_co==9: 
        if ma_co==7 or ma_co==5 or ma_co==11 or ma_co==10:
            return(1)
        else:return(0)
    elif fe_co==10: 
        if ma_co==2 or ma_co==6 or ma_co==8 or ma_co==10:
            return(1)
        else:return(0)
    elif fe_co==11:  
        if ma_co==3 or ma_co==9 or ma_co==7 or ma_co==10:
            return(1)
        else:return(0)
    elif fe_co==12:  
        if ma_co==4 or ma_co==6 or ma_co==10 or ma_co==10:
            return(1)
        else:return(0)

for i in range(0,50):
    he = np.random.randint(155,185)
    we = np.random.randint(50,90)
    fe_bl = np.random.randint(1,3)
    fe_co = np.random.randint(1,12)
    ma_bl = np.random.randint(1,3)
    ma_co = np.random.randint(1,12)
    
    if he > 175:
        if we>60 and we<80:
            if fe_bl == ma_bl:
                li = CONS(fe_co, ma_co)
                if li==1:
                    like.append(1)
                else:like.append(0)
            else:like.append(0)
        else:like.append(0)
    else:like.append(0)
    
    height.append(he)
    weight.append(we)
    female_blood.append(fe_bl)
    female_constell.append(fe_co)
    male_blood.append(ma_bl)
    male_constell.append(ma_co)
    
d = { "Height": height, 
     "Weight": weight,
    "female_blood": female_blood,
    "male_blood": male_blood,
    "female_constell": female_constell,
    "male_constell": male_constell,
    "like":like}
df = pd.DataFrame(d)
df
```

```python
df_train_y = df["like"].values
df_train = df.drop("like", axis=1)
df_train.head()
```

```python
dtree=DecisionTreeClassifier()
dtree.fit(df_train, df_train_y)

dot_data = StringIO()
export_graphviz(dtree, 
                out_file=dot_data,  
                filled=True, 
                feature_names=list(df_train),
                class_names=['nolike','like'],
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_pdf("tree.pdf")
```
